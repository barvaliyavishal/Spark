```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("CustSpend")
sc = SparkContext(conf = conf)
```

```python
def parseline(line):
    fields = line.split(",")
    cust = int(fields[0])
    spent = float(fields[2])
    return (cust, spent)
```

```python
rdd = sc.textFile("customer-orders.csv")
parsedOrders = rdd.map(parseline)
result = parsedOrders.reduceByKey(lambda x,y:x+y).sortByKey()
```

```python
result.collect()
```

### implemt sort by max amount spent

```python
flippedrdd = result.map(lambda x : (x[1],x[0]))
```

```python
swappedres = flippedrdd.sortByKey(ascending=False)
```

```python
swappedres.collect()
```
